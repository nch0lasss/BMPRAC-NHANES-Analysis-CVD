## 4. Background & Rationale

Cardiovascular disease (CVD) is the leading cause of mortality worldwide, with over 17 million deaths annually, and is responsible for nearly 1 in 4 deaths in the United States. The economic and societal burden is also substantial ‚Äî the CDC estimates that CVD costs the U.S. healthcare system over $200 billion per year in services, medications, and lost productivity. Importantly, a large proportion of cardiovascular events are preventable through early detection and lifestyle modification, which highlights the importance of identifying reliable biomarkers and modifiable risk factors.

CVD is influenced by both **non-modifiable factors (e.g., age, sex, genetics)** and **modifiable lifestyle factors such as diet, physical activity, sleep quality, smoking habits, and body composition.** These 7 predictors can also be classified under **categorical data(gender, smoking status, physical activity)**,  and **continuous data(age, BMI, systolic BP, sleep duration)**.

Increasingly, research attention has shifted toward preventative strategies ‚Äî especially those that can be implemented in the general population before the onset of clinical disease. In this context, biomarkers serve as early indicators of disease development and progression. They allow clinicians and researchers to detect underlying physiological changes that precede overt symptoms, offering a window of opportunity for intervention.

This study focuses on three biomarkers that reflect key aspects of cardiovascular physiology:

- **Triglycerides** ‚Äì Elevated triglyceride levels are associated with dyslipidemia, metabolic syndrome, and increased risk of atherosclerosis. High triglycerides are a strong predictor of coronary artery disease, especially in individuals with insulin resistance. Lifestyle interventions such as exercise and dietary fat reduction can significantly lower triglyceride concentrations.

- **Glycated Hemoglobin (HbA1c)** ‚Äì HbA1c reflects average blood glucose over 2‚Äì3 months and is associated with vascular damage, even at sub-diabetic levels. Elevated HbA1c has been linked to endothelial dysfunction and arterial stiffness. It is responsive to physical activity, diet, sleep, and weight loss.

- **Neutrophil-to-Lymphocyte Ratio (NLR)** ‚Äì A marker of systemic inflammation, calculated from white blood cell counts. Elevated NLR has been associated with coronary syndromes, heart failure, and stroke, and serves as a sensitive indicator of inflammation.

### NHANES as a Data Source

The National Health and Nutrition Examination Survey (NHANES) is conducted in 2-year cycles by the CDC and combines interviews, exams, and lab tests across a representative U.S. population. It is ideal for this type of biomedical analysis. We selected six NHANES cycles (2007‚Äì2018) for comprehensive temporal coverage.

### Supporting Literature

- **Lee & Taylor (2020)**: Showed moderate-to-vigorous physical activity correlated with lower triglycerides and higher HDL cholesterol. However, they examined only one lifestyle factor and excluded HbA1c/NLR.
- **Jones & Smith (2021)**: Found high BMI and poor sleep associated with elevated NLR/CRP, but did not include glucose or lipid markers.  

Our study addresses these gaps by analyzing three diverse biomarkers, seven predictors, and potential interaction effects over 10+ years of data.

## üì¶ Importing Libraries and Configuration

This section sets up the environment and configurations for processing NHANES data. The pandas library, which is essential for data manipulation and analysis, is imported. Additionally, the dictionaries bellow are defined and help store the following data:

- `years_cycles`: Maps NHANES years to their file cycle codes (e.g. the 2007 - 2008 cycle has the letter code 'E')
- `file_prefixes`: Maps data domains (e.g., DEMO, BMX) to their filenames.
- `sleep_var_map`: Adjusts for changes in variable names across cycles, maps each cycle code to the correct variable.
- `column_map`: Specifies which variables to extract from each dataset.
- `cvd_keywords`: List of medications related to cardiovascular diseases.

```python
# Importing pandas library to handle dataframes and tabular data
import pandas as pd

# Dictionary to define the NHANES cycle and map to their letter codes
years_cycles = {
    "2007": "E", "2009": "F", "2011": "G",
    "2013": "H", "2015": "I", "2017": "J"
}

# Dictionary of files to be downloaed from NHANES
file_prefixes = {
    'DEMO': 'DEMO', 'BMX': 'BMX', 'BPX': 'BPX', 'SLQ': 'SLQ',
    'SMQ': 'SMQ', 'PAQ': 'PAQ', 'TRIGLY': 'TRIGLY',
    'CBC': 'CBC', 'GHB': 'GHB', 'MCQ': 'MCQ'
}

# Dictionary to map the variable name for sleep hours a it differs by cycle, maps each cycle code to the correct variable
sleep_var_map = {
    "E": "SLD010H", "F": "SLD010H", "G": "SLD010H", "H": "SLD010H",
    "I": "SLD012", "J": "SLD012"
}

# Specifies which columns (ie. variables required) to retain from each file to simplify downstream merging and processing
column_map = {
    'DEMO': ['SEQN', 'RIDAGEYR', 'RIAGENDR'],
    'BMX': ['SEQN', 'BMXBMI'],
    'BPX': ['SEQN', 'BPXSY1', 'BPXSY2', 'BPXSY3'],
    'SMQ': ['SEQN', 'SMQ020', 'SMQ040'],
    'PAQ': ['SEQN', 'PAQ650'],
    'TRIGLY': ['SEQN', 'LBXTR'],
    'CBC': ['SEQN', 'LBDNENO', 'LBDLYMNO'],
    'GHB': ['SEQN', 'LBXGH'],
    'MCQ': ['SEQN', 'MCQ160B', 'MCQ160C', 'MCQ160D', 'MCQ160E', 'MCQ160F']
}

# List of keywords for CVD-related medications
cvd_keywords = [
    "statin", "aspirin", "clopidogrel", "atorvastatin", "simvastatin", "rosuvastatin",
    "beta-blocker", "metoprolol", "atenolol", "bisoprolol", "carvedilol",
    "ace", "lisinopril", "ramipril", "enalapril",
    "arb", "losartan", "valsartan", "irbesartan",
    "nitroglycerin", "isosorbide",
    "diuretic", "furosemide", "hydrochlorothiazide",
    "calcium channel blocker", "amlodipine", "verapamil", "diltiazem",
    "anticoagulant", "warfarin", "apixaban", "rivaroxaban", "dabigatran"
]

# Print success message
print("‚úÖ Libraries imported successfully.")
```

## üì• Data download and merging

This block of code loops through each NHANES cycle, downloading and reading files the relevant files specified in the dictionaries in the previous block of code.

- Extracts relevant variables and merges them across domains.
- Computes new features like Average Systolic Blood Pressure (AVG_BPXSY) and neutrophil-to-lymphocyte ratio (NLR).
- Adds the cycle lables and tracks missing data for each variable of interest.
- Stores the processed data for each cycle.
- Prepares the data for merging with medication data

```python
# Intialising of containers to store the different data
df_list = []                    # Stores the cleaned cycle datasets
dropoff_summary = {}            # Stores the dropoff stats for each cycle/variable
rx_list = []                    # Stored data on medication taken by participants
missing_data_summary = []       # Stores data on missing data counts
total_rows = 0

# Main loops that loops over each NHANES cycle
for year, cycle in years_cycles.items():
    try:
        # Downloading each specified file (e.g. DEMO, BMX) for that cycle using the corresponding URL
        dfs = {}
        for key, prefix in file_prefixes.items():
            url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{prefix}_{cycle}.XPT"
            dfs[key] = pd.read_sas(url)

        # Retrieving the correct column name for sleep data for this cycle since variable names are different for each cycle
        sleep_var = sleep_var_map[cycle]
        # Validating and filtering sleep data, keep only participants with sleep data available
        SLQ_cols = ['SEQN', sleep_var]
        if all(col in dfs['SLQ'].columns for col in SLQ_cols):
            dfs['SLQ'] = dfs['SLQ'][SLQ_cols].copy()
        else:
            continue

        # Subset columns for each dataset, retaining only necessary columns for analysis
        for key, cols in column_map.items():
            dfs[key] = dfs[key][[col for col in cols if col in dfs[key].columns]]

        # Tracking of row counts
        # Recording how many participants are in each file to later track data loss
        cycle_summary = {}
        for k in dfs:
            if 'SEQN' in dfs[k].columns:
                cycle_summary[k] = dfs[k]['SEQN'].nunique()
                cycle_summary['Unique_DEMO'] = dfs['DEMO']['SEQN'].nunique()
        cycle_summary['SLQ'] = dfs['SLQ']['SEQN'].nunique()
        dropoff_summary[year] = cycle_summary

        # Inner join all dataframes on SEQN so only participants present in all datasets are kept
        df_filtered = dfs['DEMO']
        for key in ['BMX', 'BPX', 'SLQ', 'SMQ', 'PAQ', 'TRIGLY', 'CBC', 'GHB', 'MCQ']:
            df_filtered = df_filtered.merge(dfs[key], on='SEQN', how='inner')
        dropoff_summary[year]['Merged'] = df_filtered['SEQN'].nunique()

        # Block of code to calculate the average systolic blood pressure and neutrophil-lymphocyte ratio
        df_filtered['AVG_BPXSY'] = df_filtered[['BPXSY1', 'BPXSY2', 'BPXSY3']].mean(axis=1)
        df_filtered.drop(columns=['BPXSY1', 'BPXSY2', 'BPXSY3'], inplace=True)
        df_filtered.rename(columns={sleep_var: 'Sleep_Hours'}, inplace=True)
        df_filtered['NLR'] = df_filtered['LBDNENO'] / df_filtered['LBDLYMNO']
        # Adding year identifier for reference
        df_filtered['Cycle'] = year

        # Track only missing data in final variables (excluding SEQN, Cycle, MCQ details)
        missing_vars = ['RIDAGEYR', 'RIAGENDR', 'BMXBMI', 'AVG_BPXSY',
                        'Sleep_Hours', 'SMQ020', 'SMQ040', 'PAQ650', 'LBXTR', 'NLR', 'LBXGH']
        existing_vars = [col for col in missing_vars if col in df_filtered.columns]
        # Tracking of missing data, count missing values for key variables and increment total participant count
        missing_data_summary.append(df_filtered[existing_vars].isna().sum())
        total_rows += df_filtered.shape[0]

        # Loading RX medication data and storing it for later processing
        url_rx = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/RXQ_RX_{cycle}.XPT"
        rx_df = pd.read_sas(url_rx)
        rx_df['Cycle'] = year
        rx_list.append(rx_df)

        # Storing cleaned and merged dataset for this cycle
        df_list.append(df_filtered)
        # Printing a success message to confirm all files have been loaded for that year
        print(f"‚úÖ Loaded and merged all files for {year} (Cycle {cycle}) including RXQ_RX data")

    except Exception as e:
        print(f"‚ùå Failed to process {year} (Cycle {cycle}): {e}")

# Print success message
print("‚úÖ Data download and merging completed.")
```

```python
# Combine data from all cycles into two master datasets: one for survey data and one for medications
df_rx = pd.concat(rx_list, ignore_index=True)
df_rx = df_rx[[col for col in ['SEQN', 'RXDDRUG'] if col in df_rx.columns]]
df_combined = pd.concat(df_list, ignore_index=True)

# Compute and display both counts and percentages of missing data for final key variables
total_missing = pd.DataFrame(missing_data_summary).sum().sort_values(ascending=False)
missing_percentage = (total_missing / total_rows * 100).sort_values(ascending=False)
missing_summary = pd.DataFrame({
    'Missing Count': total_missing.astype(int),
    'Missing %': missing_percentage.round(2)
})
print("\nüß™ Missing data summary for final dataset variables across all cycles:")
print(missing_summary)

# Print success message
print("‚úÖ Missing data summary computed.")
```

## üíä RX medication processing

This block of code processes prescription medication data to identify the usage of drugs that are realted to cardiovascular disease (CVD).

- Filters and normalizes medication names by converting them to lowercase.
- Identifies medications related to cardiovascular disease using the list of key words (cvd_keywords) previously defined in the "Imports and Config" section.
- Aggregates all medications per participant.
- Merges medication data with the combined dataset.

```python
# Processing RX medication data
# Converting to lowercase and filtering for CVD-related medications using keyword matching
df_rx = df_rx.dropna(subset=['RXDDRUG']).copy()
df_rx['RXDDRUG'] = df_rx['RXDDRUG'].apply(lambda x: x.decode('utf-8').lower() if isinstance(x, bytes) else str(x).lower())
df_rx = df_rx[df_rx['RXDDRUG'].str.contains('|'.join(cvd_keywords), na=False)]

# Aggregating multiple medications into a single semicolon-separated string per participant
df_rx_agg = df_rx.groupby('SEQN')['RXDDRUG'].apply(lambda x: '; '.join(x.unique())).reset_index()
# Merging medication data into the main dataset
df_combined = df_combined.merge(df_rx_agg, on='SEQN', how='left')

# Print success message
print("‚úÖ RX medication processing completed.")
```

## üßπ Data filtering and feature engineering


Cleans and enhances the merged dataset:

- Filters to include adults (participants aged ‚â• 20).
- Converts coded values to readable labels (e.g. gender codes).
- Creates a smoking status variable (Never, Former, Current).
- Converts physical activity to Yes/No.
- Creates a CVD diagnosis indicator based on answers in the questionare section of NHANES.
- Drops rows with missing values in essential columns.

```python
# Filtering participants by age
pre_filter_size = df_combined.shape[0]
# Filtering to include only adults 20 or older
df_combined = df_combined[df_combined['RIDAGEYR'] >= 20]
post_filter_size = df_combined.shape[0]

# Printing the counts of participants before and after filtering for age
print(f"\nüîç Sample size BEFORE filtering for age >= 20: {pre_filter_size}")
print(f"üîç Sample size AFTER filtering for age >= 20: {post_filter_size}")

# Create user-friendly labels for gender using a helper function
df_combined['RIAGENDR'] = df_combined['RIAGENDR'].replace({1: 'Male', 2: 'Female'})

def classify_smoking(row):
    if row['SMQ020'] == 2:
        return 'Never smoker'
    elif row['SMQ020'] == 1:
        return 'Current smoker' if row['SMQ040'] == 1 else 'Former smoker'
    return pd.NA

# Create user-friendly labels for smoking status using a helper function
df_combined['Smoking_status'] = df_combined.apply(classify_smoking, axis=1)
# Drop raw smoking variables after classification
df_combined.drop(columns=['SMQ020', 'SMQ040'], inplace=True)

# Convert physical activity indicator to readable form
df_combined['PAQ650'] = df_combined['PAQ650'].replace({1: 'Yes', 2: 'No'})

# Create a single CVD diagnosis flag based on multiple medical condition codes
cvd_cols = [col for col in ['MCQ160B', 'MCQ160C', 'MCQ160D', 'MCQ160E', 'MCQ160F'] if col in df_combined.columns]
df_combined['CVD_Diagnosed'] = df_combined[cvd_cols].apply(lambda row: any(row == 1), axis=1)

# Final clean up
# List containing names of variables relevant to final analysis
final_cols = ['SEQN', 'Cycle', 'RIDAGEYR', 'RIAGENDR', 'BMXBMI', 'AVG_BPXSY',
              'Sleep_Hours', 'Smoking_status', 'PAQ650', 'LBXTR', 'NLR', 'LBXGH', 'CVD_Diagnosed', 'RXDDRUG']
# Keep only variables relevant to the final analysis
df_combined = df_combined[final_cols]

# Defining essential columns
essential_cols = ['RIDAGEYR', 'RIAGENDR', 'BMXBMI', 'AVG_BPXSY',
                  'Sleep_Hours', 'Smoking_status', 'PAQ650', 'LBXTR', 'NLR', 'LBXGH', 'CVD_Diagnosed']

# Used to calculate the number of data points that are dropped due to NaN values
before_na = df_combined.shape[0]
# Drop rows with missing values in any key analytical variables
df_combined = df_combined.dropna(subset=essential_cols).copy()
after_na = df_combined.shape[0]

# Print success message
print("‚úÖ Data filtering and feature engineering completed.")
```

## üìà Code for dynamic descriptive statistics functions for numerical data

This block of code defines several utility functions to **analyze** and **visualize** numerical data from a DataFrame which are especially useful in **characterising** and **exploring** the identified variables. These functions include:



1.   Summary statistics to summarise the choosen numerical variable (describe_variable).
2.   Normality testing (shapiro_wilk_test).
3.   Various plots to assess distributions (plot_boxplot, plot_histogram, compare_histograms, plot_qqplot).







```python
# Importing Python libraries and functions
# pandas and numpy handle data manipulation and numerical operations
import pandas as pd
import numpy as np
# scipy.stats perform statistical tests like Shapiro-Wilk and calculate skewness
from scipy.stats import skew
import scipy.stats as stats
from scipy.stats import shapiro
# matplotlib and seaborn helps in generating plots
import matplotlib.pyplot as plt
import seaborn as sns

# Defining the function "describe_variable" for numerical data
# This function prints descriptive statistics for a numeric variable in the combined dataframe
# df is the DataFrame being analysed and column_name is the column for which you want to compute statistics
def describe_variable(df, column_name):
    # Checks and prints an error message if variable not found in the dataset
    if column_name not in df.columns:
        print(f"‚ùå Column '{column_name}' not found in the DataFrame.")
        return

    # Checks and prints an error message if the data type is not numeric
    if not pd.api.types.is_numeric_dtype(df[column_name]):
        print(f"‚ö†Ô∏è Column '{column_name}' is not numeric.")
        return

    # Removes missing values and creates a Series with all non-missing from the column
    series = df[column_name].dropna()
    # Dictionary that contains the code to compute the different descriptive statistics
    desc = {
        "Count": series.count(),
        "Minimum": series.min(),
        "25% Quartile": series.quantile(0.25),
        "Median": series.median(),
        "75% Quartile": series.quantile(0.75),
        "Maximum": series.max(),
        "Mean": series.mean(),
        "Mode": series.mode().iloc[0] if not series.mode().empty else np.nan,
        "Standard Deviation": series.std(),
        "Interquartile Range (IQR)": series.quantile(0.75) - series.quantile(0.25),
        "Skewness": skew(series)
    }

    # Prints a success message if descriptive statistics are successfuly computed
    print(f"‚úÖ Success! Descriptive statistics for '{column_name}' computed.")

    # Returns the dictionary as a Pandas Series for easy display
    return pd.Series(desc)

# Success message
print("‚úÖ Function 'describe_variable' has been defined successfully.")


# Defining the function "shapiro_wilk_test" to carry out Shapiro Wilk test
# Tests whether a numerical variable is normally distributed
# df is the DataFrame, column is the column to be tested and alpha is the significance level (default is 0.05)
def shapiro_wilk_test(df, column, alpha=0.05):
    # Checks and prints an error message if variable not found in the dataset
    if column not in df.columns:
        print(f"‚ùå Column '{column}' not found in the DataFrame.")
        return None

    # Creates a Series of non-missing values from the target column
    data = df[column].dropna()

    # Shapiro-Wilk test requires at least 3 data points, if there ar less than 3 data points, it the function exits
    if len(data) < 3:
        print("‚ùå Not enough data to perform Shapiro-Wilk test (minimum is 3 observations).")
        return None

    # Performs the Shapiro-Wilk test using scipy.stats.shapiro(), which returns the test statistic (stat) and the p-value (p)
    stat, p = shapiro(data)

    # Prints the results of the test in a readable format
    print(f"\nüìä Shapiro-Wilk Test for '{column}':")
    print(f"Test Statistic: {stat:.4f}")
    print(f"p-value: {p:.4g}")

    # Warns the user if the sample size is large
    # Scipy recommends caution with p-values for large samples
    if len(data) > 5000:
        print("‚ö†Ô∏è Note: For N > 5000, the p-value may not be accurate (per scipy warning).")

    # Intepreting the p-value
    if p > alpha:
        print("‚úÖ Fail to reject the null hypothesis. Data appears to be normally distributed.")
    else:
        print("‚ùå Reject the null hypothesis. Data does not appear to be normally distributed.")

    # Returns the test statistic and p-value for further use if needed
    return stat, p

# Success message
print("‚úÖ Function 'shapiro_wilk_test' has been defined successfully.")


# Defining the function "plot_boxplot" for numerical data
# Creates a box plot for a numerical variable in the DataFrame using Seaborn and Matplotlib
# Takes a DataFrame (df), a column name (column) to plot, and optional arguments for custom title and x-axis label
def plot_boxplot(df, column, title=None, xlabel=None):
    # Checks and prints an error message if variable not found in the dataset
    if column not in df.columns:
        print(f"‚ùå Column '{column}' not found in the DataFrame.")
        return

    # Checks and prints an error message if the data type is not numeric
    if not pd.api.types.is_numeric_dtype(df[column]):
        print(f"‚ö†Ô∏è Column '{column}' is not numeric and cannot be plotted as a boxplot.")
        return

    # Initializes a new Matplotlib figure with a size of 6 inches by 5 inches
    plt.figure(figsize=(6, 5))
    # Uses Seaborn to draw a vertical box plot of the variable
    # y=df[column] makes it vertical (box on y-axis) while color='skyblue' gives it a pleasant light-blue fill
    sns.boxplot(y=df[column], color='skyblue')

    # Sets the plot's title, uses a custom title if given
    plt.title(title if title else f'Box Plot of {column}', fontsize=14)
    # Labels the y-axis with the variable name
    plt.ylabel(column)
    # Sets the x-axis label, uses the custom label if provided
    plt.xlabel(xlabel if xlabel else "")
    # Adds a horizontal dashed grid for visual reference
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    # Adjusts plot spacing automatically to prevent overlap
    plt.tight_layout()
    # Displays the plot
    plt.show()

    # Prints a success message once the plot is generated
    print(f"‚úÖ Success! Box Plot for '{column}' computed.")

# Success message
print("‚úÖ Function 'plot_boxplot' has been defined successfully.")


# Defining the function "plot_hisotgram" to plot histograms
# Creates a histogram for a numeric variable in the DataFrame, with an optional median line
import matplotlib.pyplot as plt

def plot_histogram(df, column, bins=30, title=None, xlabel=None, ylabel='Frequency', color='skyblue', show_median=True):
    # Checks and prints an error message if variable not found in the dataset
    if column not in df.columns:
        print(f"‚ùå Column '{column}' not found in the DataFrame.")
        return

    # Drops any missing values (NaN) from the column so they don‚Äôt interfere with the histogram
    data = df[column].dropna()

    # Computes median of the variable for later plotting if show_median=True
    median_val = data.median()

    # Set default titles if not provided
    if not title:
        title = f"Histogram of {column}"
    # If no x-axis label was passed, use the column name as the label
    if not xlabel:
        xlabel = column

    # Plot histogram
    # Creates a new Matplotlib figure with a width of 8 inches and height of 5 inches
    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=bins, color=color, edgecolor='black')

    # Adds a vertical dashed red line at the median value if show_median=True
    if show_median:
        plt.axvline(median_val, color='red', linestyle='dashed', linewidth=2, label=f'Median = {median_val:.2f}')
        plt.legend()

    # Sets the plot title and axis labels
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # Adds a grid for easier reading of the histogram bars
    plt.grid(True, linestyle='--', alpha=0.7)
    # Automatically adjusts the layout so that labels and titles fit well
    plt.tight_layout()
    # Displays the histogram in the output
    plt.show()

    # Prints a success message including the median value used in the plot
    print(f"‚úÖ Histogram for '{column}' generated successfully (Median = {median_val:.2f}).")

# Success message
print("‚úÖ Function 'plot_histogram' has been defined successfully.")


# Defining the function "compare_histogram" to help visualise skewed data better
# Help visually inspect skewed data by plotting the raw histogram, a filtered histogram (1st to 99th percentile) and a log-transformed histogram
def compare_histograms(df, column, bins=30):
    # Imports NumPy and Matplotlib
    import numpy as np
    import matplotlib.pyplot as plt

    # Checks and prints an error message if variable not found in the dataset
    if column not in df.columns:
        print(f"‚ùå Column '{column}' not found in the DataFrame.")
        return

    # Removes missing values from the column to avoid errors or skewed visualizations
    data = df[column].dropna()

    # Remove zeros for log transformation
    data_log = np.log1p(data)

    # Define percentiles for outlier filtering, keeps only the values between the 1st and 99th percentiles
    # Gives a clearer view of the core distribution
    lower = data.quantile(0.01)
    upper = data.quantile(0.99)
    data_filtered = data[(data >= lower) & (data <= upper)]

    # Creates a single row of 3 plots side by side. Total width = 18 inches
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # 1. Raw histogram
    axes[0].hist(data, bins=bins, color='skyblue', edgecolor='black')
    axes[0].set_title("Raw Histogram")
    axes[0].set_xlabel(column)
    axes[0].set_ylabel("Frequency")
    axes[0].grid(True, linestyle='--', alpha=0.7)

    # 2. Filtered histogram (1st‚Äì99th percentile)
    axes[1].hist(data_filtered, bins=bins, color='orange', edgecolor='black')
    axes[1].set_title("Filtered Histogram (1st‚Äì99th percentile)")
    axes[1].set_xlabel(column)
    axes[1].set_ylabel("Frequency")
    axes[1].grid(True, linestyle='--', alpha=0.7)

    # 3. Log-transformed histogram
    axes[2].hist(data_log, bins=bins, color='green', edgecolor='black')
    axes[2].set_title("Log-Transformed Histogram (log1p)")
    axes[2].set_xlabel(f"log({column} + 1)")
    axes[2].set_ylabel("Frequency")
    axes[2].grid(True, linestyle='--', alpha=0.7)

    # Optimizes spacing so that subplots don‚Äôt overlap
    plt.tight_layout()
    # Displays the 3 histograms side by side in the output
    plt.show()

    # Prints a success message once all plots are generated
    print(f"‚úÖ Comparison histograms for '{column}' generated successfully.")

# Success message
print("‚úÖ Function 'compare_histograms' has been defined successfully.")


# Defining the function "plot_qqplot" to plot bar graphs
# Helps assess whether a variable follows a normal distribution
def plot_qqplot(df, column, dist='norm', title=None):
    # Checks and prints an error message if variable not found in the dataset
    if column not in df.columns:
        print(f"‚ùå Column '{column}' not found in the DataFrame.")
        return

    # Removes any missing values (NaN) from the column to avoid issues during plotting
    data = df[column].dropna()

    # Creates a square plotting area (6x6 inches) for a clean and readable Q-Q plot
    plt.figure(figsize=(6, 6))
    # Generates the Q-Q plot
    stats.probplot(data, dist=dist, plot=plt)
    # Uses the custom title if provided
    plt.title(title if title else f"Q-Q Plot of '{column}'")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    # Displays the Q-Q plot inline in the output
    plt.show()

    # Prints a success message to confirm that the plot has been rendered
    print(f"‚úÖ Q-Q plot for '{column}' generated successfully.")

# Success message
print("‚úÖ Function 'plot_qqplot' has been defined successfully.")
```

## üìä Code for dynamic descriptive statistics functions for categorical data

This block of code defines several utility functions to analyze and visualize categorical data from a DataFrame which are especially useful in characterising and exploring the identified variables. This includes:

1. Summary statistics to summarise the choose categorical data (describe_categorical).
2. Categorical bar chart plot to visualise participant counts in each categorey of a variable (plot_categorical_bar).

```python
# Importing Python libraries and functions
import matplotlib.pyplot as plt

# Defining the function "describe_categorical" for categorical data
def describe_categorical(df, column, unknown_codes=[7, 9, 77, 99]):
    # Checks and prints an error message if variable not found in the dataset
    if column not in df.columns:
        print(f"‚ùå Column '{column}' not found in the DataFrame.")
        return None

    # Get value counts and percentages
    # value_counts(): counts each unique value, including NaNs
    # normalize=True: returns relative frequencies (used for percentages)
    # sort_index(): ensures categories appear in numeric order
    value_counts = df[column].value_counts(dropna=False).sort_index()
    percentages = df[column].value_counts(normalize=True, dropna=False).sort_index() * 100

    # Creates a summary DataFrame of counts and percentages (rounded to 2 decimals)
    summary = pd.DataFrame({
        'Count': value_counts,
        'Percentage': percentages.round(2)
    })

    # Calculate missing and unknown counts
    total = len(df)
    missing = df[column].isna().sum()
    unknown_count = df[column].isin(unknown_codes).sum() if unknown_codes else 0

    # Displays the frequency table and total, missing, and unknown responses
    print(f"\nüìä Descriptive Statistics for '{column}':\n")
    print(summary)
    print(f"\nTotal responses: {total}")
    print(f"Missing responses: {missing}")
    if unknown_codes:
        print(f"‚ÄòDon't know‚Äô or ‚ÄòRefused‚Äô responses: {unknown_count}")

    # Returns the DataFrame so it can be saved or used later
    return summary

# Success message
print("‚úÖ Function 'describe_categorical' has been defined successfully.")


# Defining the function "plot_categorical bar" to plot bar graphs
# Creates a bar chart for a categorical variable
def plot_categorical_bar(df, column, title=None, xlabel=None):
    # Checks and prints an error message if variable not found in the dataset
    if column not in df.columns:
        print(f"‚ùå Column '{column}' not found in the DataFrame.")
        return

    # Counts the frequency of each category including missing values, sorted by index
    counts = df[column].value_counts(dropna=False).sort_index()

    # Creates a canvas for the plot, 8 inches wide by 5 inches tall
    plt.figure(figsize=(8, 5))
    bars = plt.bar(counts.index.astype(str), counts.values, color='skyblue', edgecolor='black')

    # Loops through each bar and adds the count on top, slightly above the bar
    for bar in bars:
        height = bar.get_height()
        plt.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3), textcoords="offset points", ha='center', fontsize=10)

    # Chart aesthetics
    plt.title(title if title else f'Bar Chart of {column}', fontsize=14)
    plt.xlabel(xlabel if xlabel else column)  # Use custom xlabel if provided
    plt.ylabel('Number of Participants')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    # Prevents overlaps
    plt.tight_layout()
    # Renders the plot
    plt.show()

    # Prints a message to confirm the plot was generated
    print(f"‚úÖ Success! Bar chart for '{column}' plotted.")

# Success message
print("‚úÖ Function 'plot_categorical_bar' has been defined successfully.")
```

## ü§ì Dynamic Code to compute Annual Median Trends

This block of code defines a utility function that helps to visualise the trend over time (across NHANES cycles) for the median value of a chosen numerical variable. It can be used to:

- See whether a health measure (e.g., BMI, blood pressure, sleep hours) has increased or decreased over time.
- Quickly spot changes in the population median across NHANES cycles.

```python
# Imports the matplotlib.pyplot module which is used to create plots
import matplotlib.pyplot as plt

# Defining the function "plot_annual_median_trend" which helps compute and render a line graph of the annual trends
def plot_annual_median_trend(df, variable, cycle_column='Cycle', title=None, ylabel=None):
    # Checks if the variable and cycle_column exist in the DataFrame
    # If it does not exist, it prints an error message and exits the function early
    if variable not in df.columns or cycle_column not in df.columns:
        print(f"‚ùå Required column(s) '{variable}' or '{cycle_column}' not found in DataFrame.")
        return

    # Groups the DataFrame by cycle (e.g., 2007, 2009, ...) and calculates the median of the specified variable for each year
    # reset_index() converts the result back to a regular DataFrame
    trend = df.groupby(cycle_column)[variable].median().reset_index()

    # Starts a new plot and sets the figure size to 8 inches wide and 5 inches tall
    plt.figure(figsize=(8, 5))
    # Plots the trend line
    plt.plot(trend[cycle_column], trend[variable], marker='o', color='teal', linewidth=2)

    # Loops through each point and adds a text annotation of the median value above each marker
    for i, val in enumerate(trend[variable]):
        plt.annotate(f"{val:.1f}", (trend[cycle_column].iloc[i], val), textcoords="offset points", xytext=(0, 5), ha='center')

    # Adds labels and styles
    plt.title(title if title else f"Median Annual Trend of {variable}", fontsize=14)
    plt.xlabel("NHANES Cycle Year")
    plt.ylabel(ylabel if ylabel else f"Median {variable}")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    # Displays the plot
    plt.show()

    # Prints a confirmation message that the function completed successfully
    print(f"‚úÖ Success! Annual Median Trend of '{variable}' computed and plotted.")

# Print success message
print("‚úÖ Function 'plot_annual_median_trend' has been defined successfully.")
```

## ü§ì Dynamic code to compute the Annual High Risk Trends

Bellow is a block of code that defines a function that can be used ti compute the annual trends of the percentage of participants in NHANES that are at high risk of CVD based on our choosen biomarkers.

```python
# Imprts the matplotlib.pyplot used for creating plots and charts
import matplotlib.pyplot as plt

# Defining the function "plot_annual_high_risk_trend" to plot a line plot of the % of high-risk participants per cycle
def plot_annual_high_risk_trend(df, variable, condition_fn, cycle_column='Cycle', title=None, ylabel='Percentage (%)'):
    # Ensures both variable and cycle_column exist in the DataFrame
    # If either is missing, it prints an error and exits the function
    if variable not in df.columns or cycle_column not in df.columns:
        print(f"‚ùå Required column(s) '{variable}' or '{cycle_column}' not found in DataFrame.")
        return

    # Prepares an empty list to store tuples of (year, percent_high_risk) for each NHANES cycle
    result = []

    # Groups the DataFrame by cycle year and iterates over each group (year is the year, group is the subset)
    for year, group in df.groupby(cycle_column):
        # Removes missing values from the selected variable in the current group
        valid_data = group[variable].dropna()
        if valid_data.empty:
            result.append((year, np.nan))
            continue
        # If no data is left after dropping NaNs, append NaN for this year and skip to the next cycle
        high_risk_count = valid_data.apply(condition_fn).sum()
        percent = (high_risk_count / len(valid_data)) * 100
        result.append((year, percent))

    # Converts the list of yearly results into a DataFrame for plotting
    trend_df = pd.DataFrame(result, columns=[cycle_column, 'Percentage'])

    # Plotting the trend line
    plt.figure(figsize=(8, 5))
    plt.plot(trend_df[cycle_column], trend_df['Percentage'], marker='o', color='crimson', linewidth=2)

    # Adding value labels to each point
    for i, val in enumerate(trend_df['Percentage']):
        if not np.isnan(val):
            plt.annotate(f"{val:.1f}%", (trend_df[cycle_column].iloc[i], val), textcoords="offset points", xytext=(0, 5), ha='center')

    # Graph asthetics
    plt.title(title if title else f"Annual Trend of High-Risk {variable}", fontsize=14)
    plt.xlabel("NHANES Cycle Year")
    plt.ylabel(ylabel)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    print(f"‚úÖ Success! Annual high-risk trend for '{variable}' computed and plotted.")

# Print success message
print("‚úÖ Function 'plot_annual_high_risk_trend' has been defined successfully.")
```

## Defining Dynamic Codes for each statistical tests

```python
# Dynamic code for T-test/Mann-Whitney U test function
from scipy import stats

def compare_two_groups(df, group_col, numeric_col, group1, group2, alpha=0.05):
    """Compare numeric_col between two groups in group_col.
    Automatically chooses t-test or Mann-Whitney U based on normality.

    Returns test name, statistic, and p-value.
    """
    data1 = df[df[group_col] == group1][numeric_col].dropna()
    data2 = df[df[group_col] == group2][numeric_col].dropna()

    # Normality test
    norm1 = stats.shapiro(data1)[1] > alpha
    norm2 = stats.shapiro(data2)[1] > alpha

    if norm1 and norm2:
        test_name = 'T-test'
        stat, p = stats.ttest_ind(data1, data2, nan_policy='omit')
    else:
        test_name = 'Mann-Whitney U'
        stat, p = stats.mannwhitneyu(data1, data2, alternative='two-sided')

    return {'test': test_name, 'statistic': stat, 'p_value': p}

# Success message
print("‚úÖ Function 'compare_two_groups' has been defined successfully.")

# Dynamic code for ANOVA/Kruskal-Wallis + post-hoc function
!pip install scikit-posthocs
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import scikit_posthocs as sp

def compare_multiple_groups(df, group_col, numeric_col, alpha=0.05):
    """Compare numeric_col across multiple groups in group_col.
    Chooses ANOVA or Kruskal-Wallis based on normality.
    Returns test results and post-hoc if applicable.
    """
    groups = df[group_col].dropna().unique()
    data_groups = [df[df[group_col] == g][numeric_col].dropna() for g in groups]

    # Check normality per group
    normality = [stats.shapiro(data)[1] > alpha for data in data_groups]

    if all(normality):
        # ANOVA
        formula = f'{numeric_col} ~ C({group_col})'
        model = ols(formula, data=df).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        p_val = anova_table['PR(>F)'][0]

        posthoc = None
        if p_val < alpha:
            tukey = pairwise_tukeyhsd(endog=df[numeric_col], groups=df[group_col], alpha=alpha)
            posthoc = tukey.summary()

        return {'test': 'ANOVA', 'anova_table': anova_table, 'posthoc': posthoc}
    else:
        # Kruskal-Wallis
        kw_stat, kw_p = stats.kruskal(*data_groups)
        posthoc = None
        if kw_p < alpha:
            posthoc = sp.posthoc_dunn(df, val_col=numeric_col, group_col=group_col, p_adjust='bonferroni')

        return {'test': 'Kruskal-Wallis', 'statistic': kw_stat, 'p_value': kw_p, 'posthoc': posthoc}

# Success message
print("‚úÖ Function 'compare_multiple_groups' has been defined successfully.")

# Dynamic code for Multiple Linear Regression function
import statsmodels.formula.api as smf

def linear_regression(df, outcome, predictors):
    """Run linear regression on outcome ~ predictors.
    predictors: list of variable names (include 'C(var)' for categoricals)
    Returns fitted model summary.
    """
    formula = f"{outcome} ~ " + " + ".join(predictors)
    model = smf.ols(formula, data=df).fit()
    return model.summary()

# Success message
print("‚úÖ Function 'linear_regression' has been defined successfully.")
```
